## 1. What is the overall distribution of customer ages in the dataset?

```python
df['customer_age']=pd.cut(df['age'],bins=[0,15,18,30,50,70],labels =['child','teen','young adult','adult','old'])
df.head()
```

```python
fig = px.histogram(df,x='customer_age',y='age')
fig.show()
```

## 2. How does the average purchase amount vary across different product categories?

```python
df.groupby('category')['purchase_amount_(usd)'].mean().plot(kind='bar')
```

## 3. Which gender has the highest number of purchases?

```python
df['gender'].value_counts().plot(kind= 'pie',explode = (0,0.1),autopct = '%1.1f%%')
```

## 4. What are the most commonly purchased items in each category?

```python
data = df.groupby(['category','item_purchased'])['purchase_amount_(usd)'].count()
data.plot(kind = 'bar')
```

## 5. Are there any specific seasons or months where customer spending is significantly higher?

```python
sp = df['season'].value_counts()
explode = (sp==sp.max()).astype(int)*0.1
sp.plot(kind = 'pie',explode = explode,autopct = '%1.1f%%')
```

## 6. What is the average rating given by customers for each product category?

```python
re = df.groupby('category')['review_rating'].mean()
re.plot(kind ='bar',figsize =(10,7))
plt.title('avg rating for each product')
plt.ylabel('avg rating')

for i ,value in enumerate(re):
    plt.text(i,value+0.1,f"{value:.2f}",ha= 'center' )
    
```

## 7. Are there any notable differences in purchase behavior between subscribed and non-subscribed customers?

```python
sub = df.groupby(['subscription_status'])['purchase_amount_(usd)'].mean()
sub.plot(kind='pie',explode =(0,0.1),autopct = '%1.1F%%')
plt.title('Avg spendig of subscribed and unsubscribed people')
```

## 8. Which payment method is the most popular among customers?

```python
pay_m = df.groupby('payment_method')['purchase_amount_(usd)'].count()
explode = (pay_m==pay_m.max()).astype(int)*0.1
pay_m.plot(kind='pie',explode= explode,figsize = (10,7),autopct = "%1.1f%%")
plt.title("most popular payment method")
```

## 9. Do customers who use promo codes tend to spend more than those who don't?

```python
ra = df.groupby('promo_code_used')['purchase_amount_(usd)'].mean()
ra.plot(kind = 'bar')
plt.title('avg spending who uses and dont use of redeem codes')
for i,value in enumerate(ra):
    plt.text(i,value+0.1,f"{value:2f}",ha='center')
```

## 10. How does the frequency of purchases vary across different age groups?

```python
pivot_table= pd.pivot_table(df,index='customer_age',columns= 'frequency_of_purchases',aggfunc='size',fill_value =0,observed=False)
pd.pivot_table(df,index='customer_age',columns= 'frequency_of_purchases',aggfunc='size',fill_value =0,observed=False)
```

```python
print("The maximum frequency from each age group is")
hn = pivot_table.idxmax(axis=1)   
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.scatter(hn.index, hn.values)
plt.title('Index of Maximum Value in Each Row')
plt.xlabel('Row Index')
plt.ylabel('Column Name')
plt.show()
```
